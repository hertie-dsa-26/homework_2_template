What does this code output? Explain **why** it produces that output, using the concepts of pointers and references in Python.

````python
def check_array(input):
    for idx in range(len(input)):
        if input[idx][0] == "1":
            input[idx] = None
    return input

original_array = ["1_3", "5_2"]
new_array = check_array(original_array)

print(original_array)
print(new_array)
````

Python variables store pointers/references to objects in memory. When we use 'original_array' as 'input' in the check_array function, 'input' becomes just another label pointing to the same object. 'input' returned is: [None, "5_2"]. This change happens to the same list that 'original_array' points to; the function essentially returns the modified list. Then, 'new_array' is created to point to the same modified list. Therefore, this code outputs: [None, "5_2"] [None, "5_2"]
